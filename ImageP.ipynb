```
#Reading images
from google.colab.patches import cv2_imshow #Open cv import
def read(fname):
  image = cv2.imread(fname)
  cv2_imshow(image)
```

```
read("pic1.jpg")
```

```
#Resizing image

def resize(fname,width,height):
  image = cv2.imread(fname)
  org_height , org_width = image.shape[0:2]
  print("Width: ", org_width)
  print("Height: ",org_height)
  if org_width > org_height:
    new_image = cv2.resize(image,(width,height))
  else:
    new_image = cv2.resize(image,(height,width))
  return new_image
```

```
new_image = resize("pic1.jpg",1000,1260)
cv2_imshow(new_image)
```

```
#Bluring Image
def blur(fname):
  image = cv2.imread(fname)
  kernels = [3,5,9,13]
  for k in kernels:
    image_bl = cv2.blur(image,ksize = (k,k))
    cv2_imshow(image_bl)
```

```
blur("pic1.jpg")
```

```
#Sharpening Image
import numpy as np
def sharpen(fname):
  image = cv2.imread(fname)
  kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
  new_image = cv2.filter2D(image,-1,kernel)
  cv2_imshow(new_image)
```

```
sharpen("pic1.jpg")
```
